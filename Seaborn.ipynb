# Seaborn Tutorial
# Distribution plots

```python
import seaborn as sb
```

```python
df= sb.load_dataset('tips')
df.head()  #here tip is a dependent feature on total bill, sex, smoker which are independent features
# print(type(df))
```

# Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

correlation can be found when other features are int or float, correlation cant be found with categorical features which is object (string)

```python
df.dtypes  #only float or integer vale can have correlation
```

```python
df.corr(numeric_only=True)  #numeric_only=True other wise it will through error could not convert string to float: 'No'b
# print(type(df.corr(numeric_only=True)))
```

```python
sb.heatmap(df.corr(numeric_only=True))
```

# Joinplot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot
# Univariate Analysis

```python
sb.jointplot(df,x='tip',y='total_bill',kind='hex') #histogram also showed, default kind=scatter

sb.jointplot(df,x='total_bill',y='tip',kind='hex') #histogram also showed, default kind=scatter
#both same graph
```

```python
sb.jointplot(df,x='tip',y='total_bill',kind='reg',color='cyan')
```

```python
sb.pairplot(df)
```

```python
df['smoker'].value_counts()
```

```python
sb.pairplot(df,hue='smoker',hue_order=['Yes','No'])
```

# Dist Plot
- Dist plot helps us to check the distribution of the columns feature, histrogram

```python
sb.displot([df['tip'],df['total_bill']],kde=True)
sb.displot([df['tip']],kde=True)  #it show the count, 
sb.displot([df['tip']],kind='kde') #kind=kde shows density, left axis can interprete as percentage
```

# Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

- boxplot
- violinplot
- countplot
- barplot

```python
## count plot
# one axis will always give count
# sb.countplot('sex',df) #
# sb.countplot(data=df,x='sex') #sns.countplot(titanic, x="class", hue="survived", stat="percent")
# # sb.countplot(data=df,x='day',hue='smoker')
sb.countplot(data=df,y='day',hue='smoker',palette ={'Yes':'gold','No':'cyan'}) 
# sb.countplot(data=df,y='day',hue='sex') 
```

```python
# bar plot
# need to mentioned x and y axis

# sb.barplot(df,x='total_bill',y='sex')
# sb.barplot(df,x='total_bill',y='smoker')
sb.barplot(df,y='total_bill',x='smoker')

sb.barplot(df,x='day',y='tip',hue='smoker',palette='rainbow')
```

```python
# Box plot
# percentile plot

sb.boxplot(df,x='day',y='total_bill',palette='rainbow')
```

```python
# sb.boxplot(df,x='day',y='total_bill',hue='smoker',orient='v')
sb.boxplot(df,x='total_bill',y='day',hue='smoker',orient='h')
```

# Violin Plot
Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sb.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
