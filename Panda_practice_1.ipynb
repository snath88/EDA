```python
df_mo=pd.read_excel('movies_db.xlsx',sheet_name='movies')  #read excel not csv
df_mo.head(10)
```

```python
#convert the $$ and Dollars to USD by using converter function; converters={'column_name':custom function name}

def convert(curr):
    if curr=='$$' or curr=='Dollars':
        return 'USD'
    else:
        return curr

df_fin=pd.read_excel('movies_db.xlsx',sheet_name='financials',converters={'currency':convert})
df_fin.head(10)
```

```python
#merge 2 data frame by single identical column
# >>> df1
#     lkey value
# 0   foo      1
# 1   bar      2
# 2   baz      3
# 3   foo      5
# >>> df2
#     rkey value
# 0   foo      5
# 1   bar      6
# 2   baz      7
# 3   foo      8

# Merge df1 and df2 on the lkey and rkey columns. The value columns have
# the default suffixes, _x and _y, appended.
# >>> df1.merge(df2, left_on='lkey', right_on='rkey')
df_merge=pd.merge(df_mo, df_fin, on='movie_id')
df_merge
```

```python
#convert day column to date type
# parse_dates : bool, list of Hashable, list of lists or dict of {Hashable : list}, default False
#     The behavior is as follows:

#     * ``bool``. If ``True`` -> try parsing the index.
#     * ``list`` of ``int`` or names. e.g. If ``[1, 2, 3]`` -> try parsing columns 1, 2, 3
#       each as a separate date column.
#     * ``list`` of ``list``. e.g.  If ``[[1, 3]]`` -> combine columns 1 and 3 and parse
#       as a single date column.
#     * ``dict``, e.g. ``{'foo' : [1, 3]}`` -> parse columns 1, 3 as date and call
#       result 'foo'
# df=pd.read_csv('weather_data.csv',parse_dates={'date':['day']})
df=pd.read_csv('weather_data.csv',parse_dates=['day']) #index_col=0
df
df.set_index('day',inplace=True)
df
```

```python
#replace Nan with 0
df.fillna(0,inplace=True)
df
```

```python
#replace Nan value with mean or median
df=pd.read_csv('weather_data.csv',parse_dates=['day'],index_col=0) #index_col=0, set the day column as index
print(df['temperature'].median())

#dictionary to replace value according to column
values= {
    'temperature':df['temperature'].median(),
    'windspeed': df['windspeed'].median(),
    'event':'Not'
}
df.fillna(values)  #value=values optional
```

```python
#replace outliers value with Nan
df=pd.read_csv('weather_data-1.csv',parse_dates=['day'],index_col=0)
df
# df.replace(-99999,np.nan) #replace -99999 by np.nan
# df.replace([-99999,-88888],np.nan) #replace -88888 and -99999 by np.nan

#specify vale in columns by dictionary

values={
    'temperature':[-99999],    #for single vale [] may not be mentioned like [-99999] can be mentioned as -99999
    'windspeed':[-99999,-88888],
    'event':['No Event']
}

df.replace(values,np.nan,inplace=True)
df
```

```python
# Group by custom group[key] by temperature range by creating custom function
def grouper(df,idx,col):
    if 80<=df[col][idx]<=90:  #df[col].loc[idx]
        # print(idx)
        return '80-90'
    elif 50<=df.loc[idx,col]<=60:
        # print(idx)
        return('50-60')
        
    else:
        # print(idx)
        return('Others')  
        
    
g=df.groupby(lambda idx: grouper(df,idx,'temperature'))  #idx will iterate thorugh the rows of dataframe and giving the index of that
for group,data in g:
    print(group)
    print(data)
    print('\n')
```
