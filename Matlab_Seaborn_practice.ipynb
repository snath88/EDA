```python
import pandas as pd
```

```python
df_sales=pd.read_excel('linechart.xlsx')
df_sales
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(12,4))   #12,4 size is ideal to fir
plt.plot(df_sales['Quarter'],df_sales['Fridge'],color='red',label='Fridge')
plt.plot(df_sales['Quarter'],df_sales['Dishwasher'],color='blue',label='Dishwasher')
plt.plot(df_sales['Quarter'],df_sales['Washing Machine'],color='gold',label='Washing Machine') #labe to show legend
plt.title('Product Sales')
plt.ylabel('Quater')
plt.xlabel('Revenue')
plt.legend()
plt.show()   #remove the extra line
```

```python
total_sales=df_sales[['Fridge','Dishwasher','Washing Machine']].sum()
type(total_sales),total_sales, total_sales.index
```

```python
plt.pie(total_sales,labels=total_sales.index,autopct='%1.2f%%',explode=[0.1,0,0],shadow=True,startangle=140)
plt.plot()
```

```python
plt.bar(df_sales['Quarter'],df_sales['Fridge'],label='Fridge')
plt.bar(df_sales['Quarter'],df_sales['Dishwasher'],color='red',label='Dishwasher')
plt.legend()
```

```python
df_sales.plot(kind='bar',x='Quarter')  #y=['Fridge','Dishwasher']
plt.xticks(rotation=45) #rotate the x axis legends
plt.show()
```

```python
df_sales2=df_sales.set_index('Quarter')
df_sales2
```

```python
df_sales2.plot(kind='bar') ##no need to proivde seperate x axis, it will take the index as x asis , i.e. quarter
plt.xticks(rotation=45)
plt.show()
```
