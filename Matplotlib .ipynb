```python
import matplotlib.pyplot as plt
# %matplotlib inline #plot.show()
```

```python
import numpy as np
```

```python
x= np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,s=5,c='g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph in 2D')
plt.savefig('Test1.png')
# plt.show()
```

```python
# plt.plot(x,y,'r--')   #plt.plot(x,y,'ro') #scatter line
# y=x*x
plt.plot(x,y,color='green', marker='*', linestyle='solid', linewidth=2, markersize=10)  #marker='-' dashed
plt.xlabel('X Axis') 
plt.ylabel('Y Axis')
plt.title('Sample plot')
# >>> plot(x, y)        # plot x and y using default line style and color
# >>> plot(x, y, 'bo')  # plot x and y using blue circle markers
# >>> plot(y)           # plot y using x as index array 0..N-1
# >>> plot(y, 'r+')     # ditto, but with red plusses
```

```python
plt.subplot(2,2,1)   #row, column, position
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y,'--g')
plt.subplot(2,2,2)   #row, column, position
plt.plot(x,y,'--r')
plt.subplot(2,2,3)   #row, column, position
plt.plot(x,y,'--b')
plt.subplot(2,2,4)   #row, column, position
plt.plot(x,y,'--c')
plt.savefig('test1.png')
```

```python
x=np.arange(0,4*np.pi,0.1)
# x=np.arange(0,40,0.1)
print(type(x))
print(x)
y=np.sin(x)
# print(y)

plt.plot(x,y)
plt.title('Sign curve')
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)  #2 rows 1 column, 1st index position
plt.plot(x,y_sin,'r--')
plt.title('Sign plot')
plt.subplot(2,1,2)
plt.title('Cos plot')
plt.plot(x,y_cos,'g--')
```

```python
#bar graph
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
```

```python
# Histograms
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
print(type(a))
a1=np.arange(0,8*np.pi,1)
print(a1)

# plt.hist(a1)   #x asis is the value and y axis is the density of the value, bins=10 default, density of number between 45-60 is around 2
plt.hist(a,bins=10,color='r')
```

```python
#box plot
data=[np.random.normal(0,std,100) for std in range(1,4)]
# print(data)
# print(type(data))
plt.boxplot(data,vert=True,patch_artist=True)  #percentile, 0-25-50-75-100 percentile
```
